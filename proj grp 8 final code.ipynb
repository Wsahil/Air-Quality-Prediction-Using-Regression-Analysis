```python

# Define the calculate_aqi function
def calculate_aqi(pm10, pm25, o3, no2, so2):
    # Set breakpoint concentrations
    bpm10 = 50
    bpm25 = 25
    bo3 = 120
    bno2 = 200
    bso2 = 125
    
    # Calculate pollutant indices
    ipm10 = (pm10 / bpm10) * 50
    ipm25 = (pm25 / bpm25) * 50
    io3 = (o3 / bo3) * 100
    ino2 = (no2 / bno2) * 200
    iso2 = (so2 / bso2) * 125
    
    # Calculate AQI
    aqi = round((ipm10 + ipm25 + io3 + ino2 + iso2) / 5)
    
    # Categorize AQI
    if aqi <= 750:
        category = "Good"
    elif aqi <= 900:
        category = "Satisfactory"
    elif aqi <= 1050:
        category = "Fairly polluted"
    elif aqi <= 1200:
        category = "Poor"
    else:
        category = "Hazardous"
    
    return aqi, category

# Apply the calculate_aqi function to each row of the DataFrame
df[['aqi', 'category']] = df.apply(lambda row: calculate_aqi(row['PT08.S1(CO)'], row['PT08.S2(NMHC)'], row['PT08.S3(NOx)'], row['NO2(GT)'], row['PT08.S4(NO2)']), axis=1,  result_type='expand')

# Print the updated DataFrame
df
#The most important factor in predicting the air quality index (AQI) is the concentration of pollutants in the air. The AQI is calculated based on the levels of five major pollutants: particulate matter (PM2.5 and PM10), ozone (O3), nitrogen dioxide (NO2), sulfur dioxide (SO2), and carbon monoxide (CO).

#The concentration of these pollutants is affected by a variety of factors such as weather conditions, human activities, and natural sources. For example, temperature, humidity, wind speed, and direction can affect the dispersion and accumulation of pollutants in the air. Human activities such as transportation, industry, and energy production can also contribute to the release of pollutants. Natural sources such as wildfires, dust storms, and volcanic eruptions can also affect air quality.

#Therefore, accurate measurements of the concentration of pollutants, as well as monitoring of these factors, are crucial in predicting the AQI.





```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from pmdarima.arima import *
```

```python
# Decompose the time series to identify trend, seasonality, and residuals
decomposition = seasonal_decompose(tdf1['aqi'], model='additive', period=195)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
# Test for stationarity using the Augmented Dickey-Fuller test
adf_result = adfuller(tdf1['aqi'])
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
print(f'Critical Values: {adf_result[4]}')

#In the Augmented Dickey-Fuller (ADF) test, the null hypothesis is that a time series is non-stationary, which means it has a unit root and exhibits a trend. The alternate hypothesis is that the time series is stationary, which means it has no unit root and does not exhibit a trend.

#The p-value returned by the ADF test represents the probability of obtaining a test statistic as extreme as or more extreme than the observed statistic, assuming the null hypothesis is true. If the p-value is less than the significance level (usually 0.05), then we reject the null hypothesis and conclude that the time series is stationary.

#In your case, the p-value is 0.04669451010013208, which is less than 0.05. This means that you can reject the null hypothesis and conclude that the time series is stationary with a 95% confidence level. Therefore, you can proceed with fitting a time series model on the stationary data.
```

```python
# Plot the autocorrelation and partial autocorrelation functions
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 8), sharex=True)
plot_acf(tdf1['aqi'], ax=ax[0], lags=50)
plot_pacf(tdf1['aqi'], ax=ax[1], lags=50)
plt.tight_layout()
```
